```python
df = pd.read_csv("student-mat.csv", delimiter=';')
```

```python
df.head()
```

```python
df.describe()
```

```python
# Remove students with 10 or more absences
df= df[df['absences'] < 10]
print(df.head())
```

```python
df['age'].describe()
```

```python
age_min = df['age'].min()
age_max = df['age'].max()
age_mean = df['age'].mean()
age_median = df['age'].median()
age_mode = df['age'].mode()[0]

print("Min:")
print(age_min)
print("\nMax:")
print(age_max)
print("\nMean:")
print(age_mean)
print("\nMedian:")
print(age_median)
print("\nMode:")
print(age_mode)
```

```python
# Filter the dataset to include only students from Gabriel Pereira
GP = df[df['school'] == 'GP']
print(GP.head())
```

```python
average_age_female = round(GP[GP['sex'] == 'F']['age'].mean())

print("The average age of female students from Gabriel Pereira is:", average_age_female)
```

```python
# Calculate the correlation between G1 and G3
correlation_G1_G3 = GP['G1'].corr(GP['G3'])

# Calculate the correlation between G2 and G3
correlation_G2_G3 = GP['G2'].corr(GP['G3'])

print(f"Correlation between first period grade (G1) and final grade (G3): {correlation_G1_G3}")
print(f"Correlation between second period grade (G2) and final grade (G3): {correlation_G2_G3}")
```

```python
# Determine which is more correlated
if correlation_G2_G3 > correlation_G1_G3:
    print("The second period grade (G2) is more correlated to the final grade (G3).")
else:
    print("The first period grade (G1) is more correlated to the final grade (G3).")
```
